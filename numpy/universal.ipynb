```python
import numpy as np
```

```python
array = np.array([1,2,3,4,5])
array
```

```python
#1-adição
soma = np.add(array,10)
soma
```

```python
#subtração
subtracao = np.subtract(array, 1)
subtracao
```

```python
#multiplicação
mult = np.multiply(array, 2)
mult
```

```python
#divisao
divisao = np.divide(array, 2)
divisao
```

```python
#função trigonometrica
seno = np.sin(array)
seno
```

```python
cosseno = np.cos(array)
cosseno
```

```python
media = np.mean(array)
media
```

```python
desvio_padrao = np.std(array)
desvio_padrao
```

```python
maiorq = array > 3
maiorq
```

```python
#combinando funçoes
resultado = np.sqrt(array) + np.power(array, 2)
resultado
```
